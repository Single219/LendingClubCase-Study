```python
##Aditya given hint on dividing data into bins. lets divide loan amount, annual income, interest rates into multiple bins 

?pd.cut
ld.int_rate.describe()
```

```python
##loan amount bins creation

ld['loanAmountBins'] = pd.cut(ld['loan_amnt'], [0, 6000, 12000, 18000, 24000, 35000], labels=['0-6000', '6000-12000', '12000-18000', '18000-24000', '24000+'])
```

```python
## annual income bins creation
ld['annualIncomeBins'] = pd.cut(ld['annual_inc'], [0, 20000, 40000, 60000, 80000, 100000], labels=['0-20000', '20000-40000', '40000-60000', '60000-80000', '80000+'])
```

```python
## Interate rate bins creation
ld['interestRatetBins'] = pd.cut(ld['int_rate'], [0, 5, 10, 15, 20, 25], labels=['0-5', '5-10', '10-15', '15-20', '20+'])
```

```python
#Annual income and chargedoff

annualincomevsChargedoff = ld.groupby(['annualIncomeBins', 'loan_status']).loan_status.count().unstack().reset_index()
annualincomevsChargedoff['Total'] = annualincomevsChargedoff['Charged Off'] + annualincomevsChargedoff['Current'] + annualincomevsChargedoff['Fully Paid'] 
annualincomevsChargedoff['Chargedoff_Proportion'] = annualincomevsChargedoff['Charged Off'] / annualincomevsChargedoff['Total']
annualincomevsChargedoff.sort_values('Chargedoff_Proportion', ascending=False)

fig = plt.figure(figsize =(25, 10))
ax1=sns.barplot(x='annualIncomeBins', y='Chargedoff_Proportion', data=annualincomevsChargedoff)
ax1.set_title('Annual income and Chargedoff')
plt.show()

```

```python
#Loan Amount and chargedoff

annualincomevsChargedoff = ld.groupby(['loanAmountBins', 'loan_status']).loan_status.count().unstack().reset_index()
annualincomevsChargedoff['Total'] = annualincomevsChargedoff['Charged Off'] + annualincomevsChargedoff['Current'] + annualincomevsChargedoff['Fully Paid'] 
annualincomevsChargedoff['Chargedoff_Proportion'] = annualincomevsChargedoff['Charged Off'] / annualincomevsChargedoff['Total']
annualincomevsChargedoff.sort_values('Chargedoff_Proportion', ascending=False)

fig = plt.figure(figsize =(25, 10))
ax1=sns.barplot(x='loanAmountBins', y='Chargedoff_Proportion', data=annualincomevsChargedoff)
ax1.set_title('Loan Amount and Chargedoff')
plt.show()
```

```python
#Interest Rate and chargedoff

annualincomevsChargedoff = ld.groupby(['interestRatetBins', 'loan_status']).loan_status.count().unstack().reset_index()
annualincomevsChargedoff['Total'] = annualincomevsChargedoff['Charged Off'] + annualincomevsChargedoff['Current'] + annualincomevsChargedoff['Fully Paid'] 
annualincomevsChargedoff['Chargedoff_Proportion'] = annualincomevsChargedoff['Charged Off'] / annualincomevsChargedoff['Total']
annualincomevsChargedoff.sort_values('Chargedoff_Proportion', ascending=False)

fig = plt.figure(figsize =(25, 10))
ax1=sns.barplot(x='interestRatetBins', y='Chargedoff_Proportion', data=annualincomevsChargedoff)
ax1.set_title('Interest Rate and Chargedoff')
plt.show()
```
